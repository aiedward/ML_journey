```python
import pandas as pd
import numpy as np
import seaborn as sb
```

```python
df=pd.read_csv("E:/Courses/AI/Data Science/New folder/tips.csv")
df.head()
```

# Distribution plot

### Kdeplot

```python
sb.kdeplot(df['total_bill'])
```

### Distplot

```python
sb.distplot(df['total_bill'], kde=True)
```

```python
sb.distplot(df['total_bill'], kde=False, bins=40)
```

### Jointplot

```python
sb.jointplot(df['total_bill'],df['tip'])
```

```python
sb.jointplot(df['total_bill'],df['tip'], kind='reg')
```

```python
sb.jointplot(df['total_bill'],df['tip'], kind='hex')
```

```python
sb.jointplot(df['total_bill'],df['tip'], kind='kde')
```

### Pairplot

```python
sb.pairplot(df)
```

```python
sb.pairplot(df, hue='sex')
```

# Categorial plot

### Barplot

```python
sb.barplot(df['sex'],df['total_bill'])
```

### Countplot

```python
sb.countplot(df['sex'])
```

### Stripplot

```python
sb.stripplot(x='day',y='total_bill',data=df,hue='sex',split=True, jitter=True)
```

# Matrix plot

```python
df1=df.corr()
sb.heatmap(df1)
```
